```python
url = "https://digipodium.com/mining-demo.html"
```

```python
from dputils.scrape import Scraper, Tag
```

```python
scr = Scraper(url)
```

```python
result = scr.get_repeating_date_from_page(
    target=Tag(cls='item-list'),
    items=Tag(cls="card"),
    title = Tag('h4'),
    price = Tag('h1', cls="pricing-card-title"),
    oprice = Tag('li', cls='o-price')
)
```

```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.DataFrame(result)
df
```

```python
df.price = df.price.apply(lambda p:p[1:]).astype(float)
df
```

```python
df.oprice = df.oprice.apply(lambda p:p[1:]).astype(float)
df
```

```python
import plotly.express as px
import plotly.graph_objects as go
```

```python
# Bar Chart: Prices vs. Original Prices
bar_fig = px.bar(
    df,
    x="title",
    y=["price", "oprice"],
    title="Current Prices vs. Original Prices of Products",
    labels={"value": "Price (₹)", "title": "Products"},
    barmode="group",
    template="plotly_dark",
    width=800,
    height=600,
)
bar_fig.show()
```

```python
# Pie Chart: Percentage Share of Product Prices
pie_fig = px.pie(
    df,
    values="price",
    names="title",
    title="Percentage Share of Product Prices",
    hole=0.3,
    template="plotly_dark",
    width=800,
    height=600,
)
pie_fig.show()
```

```python
# Prepare data for the bar chart race
# Add a "time" column to simulate progression
steps = 15  # Number of animation steps
frames = []

# Interpolate values over steps to simulate animation
for i in range(steps + 1):
    frame = df.copy()
    frame["step"] = i  # Add a time-like column
    frame["price_progress"] = frame["price"] * (i / steps)
    frame["oprice_progress"] = frame["oprice"] * (i / steps)
    frames.append(frame)

# Combine all frames
race_df = pd.concat(frames)

# Melt data for animation
race_df_melted = pd.melt(
    race_df,
    id_vars=["title", "step"],
    value_vars=["price_progress", "oprice_progress"],
    var_name="Price Type",
    value_name="Value",
)

# Map human-readable labels for Price Type
race_df_melted["Price Type"] = race_df_melted["Price Type"].replace(
    {"price_progress": "Price", "oprice_progress": "Original Price"}
)

# Create animated bar chart
bar_race_fig = px.bar(
    race_df_melted,
    x="Value",
    y="title",
    color="Price Type",
    orientation="h",
    animation_frame="step",
    title="Bar Chart Race: Price vs Original Price",
    labels={"Value": "Price (₹)", "title": "Product"},
    color_discrete_map={"Price": "blue", "Original Price": "red"}
)
bar_race_fig.update_layout(yaxis={"categoryorder": "total ascending"})
bar_race_fig.show()
```
